```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

tesla = yf.Ticker("TSLA")
tesla_share_price_data = tesla.history(period="10y")

tesla_share_price_data.reset_index(inplace=True)

tesla_share_price_data.plot(title="Tesla 10y Stock Price", x="Date", y="Open")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.show()
```
